# Q2.Creating an nlp document object

```python
#Q2creating an nlp document object
import spacy 
nlp = spacy.load("en_core_web_sm")
```

```python
#creating an array and storing all the sentences in sentence_list
sentence_list=[]
doc_object = nlp(doc_read)
for sentence in doc_object.sents:
    sentence_list.append(sentence)
```

```python
print(sentence_list[508])#choosing one of the sentences to perform some tokenization
```

```python
#Displaying (a)TOKEN, (b)TOKEN POS TAGS (C)TOKEN DEPENDENCY (D)EXPLAINATION OF EACH TOKEN POS

for token in sentence_list[508]:
    print(print(f" {token.text:{10}} {token.pos_:{10}} {token.dep_:{10}} {spacy.explain(token.pos_):{20}}"))
```

#  Q3Regular Expression is a set of special characters which is also used in python for matching of characters,patterns or strings and python has an in-built library funtion for that.


```python

#importing the regular expression library to perform some regex tasks
import re
```

```python
a = r"\b[aA]\w+"
print(f"words that start with  a of which total number of occurences are: {len(re.findall(a,doc_read))}")
e = r"\b[eE]\w*"
print(f"words that start with e of which total number of occurences are: {len(re.findall(e,doc_read))}")
i = r"\b[iI]\w*"
print(f"words that start with  i of which total number of occurences are: {len(re.findall(i,doc_read))}")
o = r"\b[oO]\w*"
print(f"words that start with  o of which total number of occurences are: {len(re.findall(o,doc_read))}")
u = r"\b[uU]\w*"
print(f"words that start with  u of which total number of occurences are: {len(re.findall(u,doc_read))}")
```

# 8 Phrase matching

```python
#importing phrasematcher from the spacy.matcher library
from spacy.matcher import PhraseMatcher
matcher = PhraseMatcher(nlp.vocab)

#creating an list of phrases for pattern matching
phraselist=["laugh","cry","happy","sad"]
phrasepatterns = [nlp.make_doc(word) for word in phraselist]

matcher.add("Emotions",None, *phrasepatterns)

getmatches = matcher(doc_object)
getmatches

for index, start, end in getmatches:
    stringid= nlp.vocab.strings[index]
    span= doc_object[start:end]
    print(f"Start:{start:5}|End:{end:5}|Text:{doc_object[start-5:end+5]}")
```
